##Function to Print a table

```python
#def is used to create function in python
def print_table(table):
  for i in range(1,11):
    print(f"{table} x {i}= {table*i}")

#function call
print_table(19)
```

#Fibonacci Series

```python
#Example :[0,1,1,2,3,5,.....]

def get_fibo(length):
  '''This func takes length as an argumernt
  ANd return fibonacci list'''
  fibo=[0,1]
  for i in range(length-2):
    last=fibo[-1]
    second_last=fibo[-2]
    new_num =  last+second_last
    fibo.append(new_num)
  return fibo

#function call
get_fibo(15)
```

```python
def even_odd(data):
  even = []
  odd = []
  for i in data:
    if i%2==0:
      even.append(i)
    else:
      odd.append(i)
  print('Even List',even)
  print('Even List',even)
```

#Student Record Management System (SRMS)

```python
def srms(Student_count):
  Data={"Name":[],"Age":[],"Course":[],"Gender":[],"Roll_no":[],"Address":[]}
                #dictionary syntax ^
  for i in range(Student_count):
    for j in Data:
      ans=input(f"Enter your {j}:")
      Data[j].append(ans)

    print(f"{i+1} record inserted succesfully")

  import pandas as pd
  table= pd.DataFrame(Data)
  display(Clear =True)
  table.to_excel("Student_record.xlsx")

  return table

number= int(input("Enter number of student:"))
srms(number)
```

#Wallpaper Downloader

```python
def wallpaper_install(width,length):
  url=f"https://picsum.photos/{length}/{width}"
  import requests as r
  data =r.get(url).content

  #file handling
  with open("wallpaper.jpeg","wb") as f:
    f.write(data)
  from IPython.display import Image
  display(Image("wallpaper.jpeg"))
  return url

wallpaper_install(1980,1080)
```

#Paytm SoundBox

```python
# gTTS: Google Text-to-Speech

def payment_box(amount, company='Paytm'):
    msg = f"{company}, पर. {amount}. रुपये. प्राप्त हुए।"

    from gtts import gTTS
    audio = gTTS(text=msg, lang='hi')
    audio.save("payment.mp3")

    from IPython.display import Audio, display
    display(Audio("payment.mp3", autoplay=True))

# Function call
payment_box(2000)
```
