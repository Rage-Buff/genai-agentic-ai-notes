##Inheritance

```python
#Inheritance: To pass methods or features
# from parent class to child class
#without creating it again

#Single | Simple
class Laptop:     #Super Class
  def system_on(self):
    print('System is ON!')

  def task_manager(self):
    print('Show Current Running Process!!')

class Mobile(Laptop):     #Sub Class
  def play_game(self):
    print("User can play games")
```

```python
nokia=Mobile()
nokia.system_on()
nokia.task_manager()
nokia.play_game()
```

```python
#Multiple Inheritance
class Samsung:
  def display(self):
    print('High Quality Display')
class Neo:
  def charger(self):
    print('Super fast charger')
class Sony:
  def camera(self):
    print('72 Megapixel camera')
class Phone(Samsung,Neo,Sony):
  def my_phone(self):
    print('This is a Phone')

galaxy=Phone()
galaxy.display()
galaxy.charger()
galaxy.camera()
galaxy.my_phone()
```

```python
#Multilevel Inheritance
class Grand_Father:
  def land_property(self):
    print('Huge Land Property')
class Father(Grand_Father):
  def flats(self):
    print("10+ Flats")
class Son(Father):
  def bike_collection(self):
    print('5+ Bike Collection')

class Grand_Son(Son):
  def nothing(self):
    print('Nothing phone')

vansh = Grand_Son()
vansh.land_property()
vansh.flats()
vansh.bike_collection()
vansh.nothing()
```

```python
#Hierarchical Inheritance
class Trainer:
  def python(self):
    print("I'm Learning Python")
class Harry(Trainer):
  def java(self):
    print("I know java fullstack")
class Naman(Trainer):
  def php(self):
    print("I know php fullstack")

a=Harry()
b=Naman()
a.python()
a.java()
print("=================")
b.python()
b.php()

```

```python
# 5. Hybrid: Inheritence
# Combination of any two or more

class Hybrid(Trainer,Phone,Laptop):
  def Hybrid_func(self):
    print("This is Hybrid Class")

h = Hybrid()
h.python()   # comes from Trainer class
h.camera()   # comes from Phone class
h.task_manager()   # comes from Laptop class
```

```python
##Method Overloading

class IITM:
  def introduce(self, name= None):
    if name == None:
      print("Hello student,How are you?")
    else:
      print(f"Hello {name}, how are you?")
      print("Nice to meet you!")
gaurav=IITM()
gaurav.introduce()
print("===============================")
gaurav.introduce("Gaurav")
```

```python
#Overriding
class Animal:
  def speak(self):
    print("I can speak")
class Dog(Animal):
  def speak(self):
    print("Bark! Bark!")
class Cat(Animal):
  def speak(self):
    print("Meow! Meow!")
print("=====Animal Class Object=====")
a=Animal()
a.speak()       #Same Method
print("=====Dog Class Object=====")
a=Dog()
a.speak()       #Same Method
print("=====Cat Class Object=====")
a=Cat()
a.speak()       #Same Method
```

```python
from types import MethodDescriptorType
# Abstraction: Hides Internal working and enforce
# method execution in different class

# abc: module
# ABC: Abstract Class
# abstarct method: to implement Abstraction
from abc import ABC, abstractmethod
class Website_design(ABC):
  @abstractmethod
  def login(self):
    print("User Login Form")

  @abstractmethod
  def logout(self):
    print("User Logout Form")

print('Done')
```

```python
class Intern(Website_design):
  def login(self):
    print("This is Login Page")
  def logout(self):
    print("This is Logout Page")
  def contact_us(self):
    print("This is Contact us Page")
vansh = Intern()
vansh.contact_us()
```
