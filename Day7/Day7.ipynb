#Data Analysis using Python

```python
import pandas as pd  #Table
import numpy as np   #Numerical fast array
import seaborn as sns #Advance visuals charts
import matplotlib.pyplot as plt #Basic visuals charts
import os #File path read
import warnings #future warnings ignore
warnings.filterwarnings('ignore')
print("Done")
```

```python
#Employee Data
data= {'Emp_Id':[101,102,105,106,108],
       "Name":['Ajay',"Mohan","Ravi","Karan","Jatin"],
       "Salary":np.random.randint(15000,50000,5),
       "Address":['Noida',"Delhi","Gr Noida","Pune","Noida"],
       "Dept":['HR','IT','Sales','Admin','IT']}
table =pd.DataFrame(data)
table       #table is called DataFrame
```

```python
#Q1: Check table shape, rows and columns
table.shape
#(Rows,Columns)
```

```python
#Q2: Show each employee salary using plot
x= table['Name']
y= table['Salary']
plt.title("Employee Salary",fontweight='bold')
plt.bar(x,y,color=['r','silver','b','orange','cyan'])
plt.xlabel("Employee Name",color='b')
plt.ylabel("Salary",color='r')
plt.show()
```

```python
#Q3: Show Company Department Wise salary expense
table.groupby('Dept')['Salary'].sum()
```

```python
#Q4: Show Company Department Wise salary expense using pie plot
table.groupby('Dept')['Salary'].sum().plot(kind='pie',autopct='%1.2f%%')  #autopct: shows percentage as %
plt.show()
```

```python
#Q5: Show Sample:
# it is required when we have large dataset
#
table.sample(2)
```

```python
#Q6: Increase each emp salary by 10k

table['Salary']+=10000
table
```

```python
#Q7: Update Karan emp id is 4 salary to 50k
# using loc method: loc: location

table.loc[3,'Salary']=50000
table
```

```python
#Q8 change name using iloc
#iloc : index based location
table.iloc[4,1]="Mr. Jatin Kumar"
table
```

```python
#Q9: create new column DOJ
# Date of joining
#YYYY-MM-DD: Format
table['DOJ']= ["2025-01-26",
               "2025-06-15",
               "2025-01-09",
               "2025-12-02",
               "2020-08-15",]
table
```

```python
#Q10: To Show each columns datatype
#use info,this also show non-null count
#null:==> missing data
table.info()
```

```python
#Q11: Change Date Object data type to datetime
#pd.to_datetime(): method to convert object to datetime

table['DOJ']=pd.to_datetime(table['DOJ'])
table.info()
```

```python
#Q12: get month_name or day_name from DOJ
table['Month_Name']=table['DOJ'].dt.month_name()
table['Day_Name']=table['DOJ'].dt.day_name()
table
```

```python
from matplotlib.lines import lineStyles
#Q13: plot emp DOJ and count
#before sorting

x=table['DOJ']
y=table['Emp_Id']
plt.title("Emp DOJ and count")

plt.plot(x,y, linestyle= '--', marker='o',color='red')
plt.xticks(rotation=45)
plt.xlabel("Emp DOJ")
plt.ylabel("Emp Count")
plt.show()
```

```python
from matplotlib.lines import lineStyles
#Q14: plot emp DOJ and count
#after sorting
table=table.sort_values(by='DOJ')
x=table['DOJ']
y=table['Emp_Id']
plt.title("Emp DOJ and count")

plt.plot(x,y, linestyle= '--', marker='o',color='r')
plt.xticks(rotation=45)
plt.xlabel("Emp DOJ")
plt.ylabel("Emp Count")
plt.show()
```

```python
#Q15: Reset_index
table.reset_index(drop= True, inplace= True)
table
#drop: incorrect index: drop
#inplace: show changes in original file values
```

```python
#Q16: Shows unique data in Address
table['Address'].unique()
```

```python
#nunique: show counts only
table['Address'].nunique()
```

```python
#Q17: Show nan count: missing data
table.isna().sum()        # isna: read as is-na
# If all zeroes, means no missing data
```

```python
#Q18: Create new column Bonus with missing values

table['Bonus']=[2000,None,1500,1200,None]
table
```

```python
#Q19: Show emp count by dept
#value_counts()
table['Dept'].value_counts()
```

```python
#Q20: How to fill missing values
#use fillna method

table['Bonus']=table['Bonus'].fillna(0)
#user can select the value
#Missing Bonus: 0
table
```

```python
#Q21: What is company total salary expense
#single column is called Series
#Double or more is called DataFrame

expense= table['Salary'].sum()
print(f"Total Salary expense: Rs {expense}")
```
