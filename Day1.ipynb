```python
 # List vs tuple
# list:
# allow any data type to store
# mutable: element can be changed
# list is slow in operation
# list mainly exists in python not in java, c, c++
# starts with square brac

# tuple
# allow any data type to store
# immutable: element can not be changed
# tuple is fast in operation
# tuple exists in most programming language
# start with round brac
```
